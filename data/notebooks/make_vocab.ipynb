```python
import pandas as pd
```

```python
test_data=pd.read_csv('archive/im2latex_test.csv')
train_data=pd.read_csv('archive/im2latex_train.csv')
val_data=pd.read_csv('archive/im2latex_validate.csv')
```

```python
formula_list = []
max_len = 0
for row in test_data['formula']:
    if len(str(row).strip("\n").split()) > max_len:
        max_len = len(str(row).strip("\n").split())
        print(row)
        print(max_len)
    formula_list += str(row).strip("\n").split()

for row in train_data['formula']:
    if len(str(row).strip("\n").split()) > max_len:
        max_len = len(str(row).strip("\n").split())
        print(row)
        print(max_len)
    formula_list +=  str(row).strip("\n").split()
for row in val_data['formula']:
    if len(str(row).strip("\n").split()) > max_len:
        max_len = len(str(row).strip("\n").split())
        print(row)
        print(max_len)
    formula_list +=  str(row).strip("\n").split()
itos = list(set(formula_list))
itos.append('<sos>')
itos.append('<eos>')
itos.append('<unc>')
itos.append('<pad>')
print(len(itos))
```

```python
def create_vocab(list):
    vocab = {}
    for i, token in enumerate(list):
        print(i, token)
        vocab[token] = i
    return vocab

stoi = create_vocab(itos)
```

```python

def tokenize_formula():
    output_length = 627
    modes = ['test','train','validate']

    for mode in modes:
        df = pd.read_csv('archive/im2latex_'+mode+'.csv')

        with open("archive/im2latex_"+mode+"_tokenized.txt", "w") as file:
            for _ , row  in df.iterrows():
                tokenize_formula = []
                formula = str(row["formula"]).strip("\n").split()
                
                tokenize_formula.append(stoi['<sos>'])
                for token in formula:
                    tokenize_formula.append(stoi[token])
                tokenize_formula.append(stoi['<eos>'])
                for _ in range(output_length - len(formula)):
                    tokenize_formula.append(stoi['<pad>'])
            
                file.write(row["image"] + "§" + str(tokenize_formula) + "\n")
    
```
